```python
import google.generativeai as genai


genai.configure(api_key="your api key")  #enter your api key
```

```python

topics = {
    "1": "Python",
    "2": "C++",
    "3": "Java",
    "4": "Full Stack Development"
}

def build_prompt(topic):
    return (
        f"Generate 10 high-quality technical interview questions for {topic}. "
        f"Include a mix of theoretical and coding questions, ensure the difficulty is balanced and make sure the questions will not repeated in any sequence or generation "
        f"(easy, medium, hard). Return only the questions in a numbered list format."

    )
```

```python
def generate_questions(prompt):
    model = genai.GenerativeModel('gemini-2.0-flash')
    response = model.generate_content(prompt)
    return response.text
```

```python
def main():
    print("Choose a topic to generate interview questions:")
    for key, value in topics.items():
        print(f"{key}: {value}")

    choice = input("Enter your choice (1-4): ").strip()
    topic = topics.get(choice)

    if topic:
        print(f"\nGenerating interview questions for {topic}...\n")
        prompt = build_prompt(topic)
        questions = generate_questions(prompt)
        print(questions)
    else:
        print("Invalid selection. Please try again.")

if __name__ == "__main__":
    main()
```
